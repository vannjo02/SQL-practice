### Which government form has the highest average life expectancy?



```python
%%sql

select governmentform, avg(lifeexpectancy) from country
group by governmentform
order by avg(lifeexpectancy) desc
limit 1;
```

### Calculate the surface area of the regions and sort in ascending order



```python
%%sql

select region, sum(surfacearea) as surfacearea from country
group by region
order by sum(surfacearea) asc;
```

### which government types have the lowest average life expectancy?



```python
%%sql

select governmentform, avg(lifeexpectancy) from country
group by governmentform
order by avg(lifeexpectancy) asc;
```

### What regions have a high average life expectancy?



```python
%%sql

select region, avg(lifeexpectancy) from country
group by region
order by avg(lifeexpectancy) desc;
```

### What is the most common form of government?



```python
%%sql

select governmentform, count(*) as govformfreq from country
group by governmentform
order by govformfreq desc;
```

### Which region has the highest total gnp?



```python
%%sql

select region, sum(gnp) as regiongnp from country
group by region
order by regiongnp desc;
```

### Show the 10 government forms with the largest population



```python
%%sql

select governmentform, sum(population) as totalpopulation from country
group by governmentform
order by totalpopulation desc
limit(10);
```

### List the total population of each continent and order them from smallest to largest



```python
%%sql

select continent, sum(population) as continentpop from country
group by continent
order by continentpop asc;
```

### Show the total number of countries for each region in Europe

```python
%%sql

select region, count(name) as countrycount from country
where continent = 'Europe'
group by region;
```

### List the total population of each region in Africa from lowest to highest.

```python
%%sql

select region, sum(population) as regionpop from country
where continent = 'Africa'
group by region
order by regionpop asc;
```
