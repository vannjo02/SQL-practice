## Aggregate Query Practice

```python
import warnings
warnings.filterwarnings('ignore')
from reframe import Relation
country = Relation('/home/faculty/millbr02/pub/country.csv')
```

### Which government form has the highest average life expectancy?



```python
country.groupby('governmentform').mean('lifeexpectancy').sort('mean_lifeexpectancy',ascending=False)
```

### Calculate the surface area of the regions and sort in ascending order



```python
country.groupby('region').sum('surfacearea').sort('sum_surfacearea')
```

### which government types have the lowest average life expectancy?



```python
country.groupby('governmentform').mean('lifeexpectancy').sort('mean_lifeexpectancy')
```

### What regions have a high average life expectancy?



```python
country.groupby('region').mean('lifeexpectancy').sort('mean_lifeexpectancy',ascending=False)
```

### What is the most common form of government?



```python
country.groupby('governmentform').count('name').sort('count_name',ascending=False)
```

### Which region has the highest total gnp?



```python
country.groupby('region').sum('gnp').sort('sum_gnp',ascending=False)
```

### Show the 10 government forms with the largest population



```python
country.groupby('governmentform').sum('population').sort('sum_population',ascending=False)
```

### List the total population of each continent and order them from smallest to largest



```python
country.groupby('continent').sum('population').sort('sum_population')
```

### Show the total number of countries for each region in Europe

```python
country.query('continent=="Europe"').groupby('region').count('name').sort('count_name',ascending=False)
```

### List the total population of each region in Africa from lowest to highest.

```python
country.query('continent == "Africa"').groupby('region').sum('population').sort('sum_population')
```
